```python
# Define file paths and load raw CSVs into a dictionary
file_paths = {
    "order_reviews": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\order_reviews_dataset.csv",
    "orders": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\orders_dataset.csv",
    "product_category_translation": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\product_category_name_translation.csv",
    "products": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\products_dataset.csv",
    "sellers": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\sellers_dataset.csv",
    "closed_deals": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\closed_deals_dataset.csv",
    "customers": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\customers_dataset.csv",
    "geolocation": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\geolocation_dataset.csv",
    "marketing_qualified_leads": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\marketing_qualified_leads_dataset.csv",
    "order_items": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\order_items_dataset.csv",
    "order_payments": r"D:\Data_Science\Projects\Portfolio_Projects\Marketing-Funnel-Seller-Performance-and-Revenue-Impact\Dataset\Raw\order_payments_dataset.csv",
}

tables = {}

# Iterate through paths, read CSVs, and confirm shape
for name, path in file_paths.items():
    tables[name] = pd.read_csv(path)
    print(f"{name} loaded successfully | Shape: {tables[name].shape}")
```

```python
# Define and run a comprehensive inspection function for each table
def inspect_table(df, table_name):
    # Print headers and shape
    print("=" * 80)
    print(f"TABLE: {table_name}")
    print("=" * 80)

    print("\nShape:")
    print(df.shape)

    # Print column names
    print("\nColumns:")
    print(df.columns.tolist())

    # Show first 5 rows
    print("\nFirst 5 rows:")
    display(df.head())

    # Show data types
    print("\nData types:")
    display(df.dtypes.to_frame("dtype"))

    # Calculate and display missing values and percentages
    print("\nMissing values:")
    missing = df.isna().sum().to_frame("missing_count")
    missing["missing_percent"] = (missing["missing_count"] / len(df) * 100).round(2)
    display(missing.sort_values("missing_count", ascending=False))

    # Count exact duplicates
    print("\nDuplicate rows:")
    print(df.duplicated().sum())

    # Summary statistics for numeric columns
    print("\nNumeric summary:")
    display(df.describe())

    # Summary statistics for text/categorical columns
    print("\nObject / categorical summary:")
    display(df.describe(include=["object", "str"]))
    
    # Show unique value counts per column
    print("\nUnique values per column:")
    display(df.nunique().to_frame("unique_values").sort_values("unique_values", ascending=False))

# Apply the function to all loaded tables
for name, df in tables.items():
    inspect_table(df, name)
```

```python
# Check for duplicates in primary key columns
primary_key_checks = {
    "customers": "customer_id",
    "orders": "order_id",
    "products": "product_id",
    "sellers": "seller_id",
    "marketing_qualified_leads": "mql_id",
    "closed_deals": "mql_id",
    "product_category_translation": "product_category_name"
}

for table_name, key_col in primary_key_checks.items():
    df = tables[table_name]
    print(f"{table_name}.{key_col}")
    print("Rows:", len(df))
    print("Unique:", df[key_col].nunique())
    print("Duplicates:", df[key_col].duplicated().sum())
    print("-" * 50)
```

```python
# Check foreign key relationships for orphaned records
relationship_checks = [
    ("orders", "customer_id", "customers", "customer_id"),
    ("order_items", "order_id", "orders", "order_id"),
    ("order_items", "product_id", "products", "product_id"),
    ("order_items", "seller_id", "sellers", "seller_id"),
    ("order_payments", "order_id", "orders", "order_id"),
    ("order_reviews", "order_id", "orders", "order_id"),
    ("closed_deals", "mql_id", "marketing_qualified_leads", "mql_id"),
    ("closed_deals", "seller_id", "sellers", "seller_id"),
    ("products", "product_category_name", "product_category_translation", "product_category_name")
]

for left_table, left_key, right_table, right_key in relationship_checks:
    left_values = set(tables[left_table][left_key].dropna())
    right_values = set(tables[right_table][right_key].dropna())

    # Find values in the left table that don't exist in the right table
    unmatched = left_values - right_values

    print(f"{left_table}.{left_key} → {right_table}.{right_key}")
    print("Unmatched values:", len(unmatched))
    print("-" * 60)
```

```python
# Identify invalid date formats across specific columns
date_columns = {
    "orders": [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ],
    "order_items": ["shipping_limit_date"],
    "order_reviews": ["review_creation_date", "review_answer_timestamp"],
    "closed_deals": ["won_date"],
    "marketing_qualified_leads": ["first_contact_date"]
}

for table_name, cols in date_columns.items():
    print(f"\n{table_name}")
    for col in cols:
        converted = pd.to_datetime(tables[table_name][col], errors="coerce")
        invalid_count = converted.isna().sum() - tables[table_name][col].isna().sum()
        print(f"{col}: invalid dates = {invalid_count}")
```

```python
# Check value distributions for key categorical columns
categorical_checks = {
    "orders": ["order_status"],
    "order_payments": ["payment_type"],
    "marketing_qualified_leads": ["origin"],
    "closed_deals": ["business_segment", "lead_type", "lead_behaviour_profile", "business_type"]
}

for table_name, cols in categorical_checks.items():
    print(f"\n{table_name}")
    for col in cols:
        print(f"\n{col}")
        display(tables[table_name][col].value_counts(dropna=False))
```

# Cleaning

```python
# Create a working copy of all tables for cleaning
cleaned_tables = {name: df.copy() for name, df in tables.items()}
```

```python
# Correct spelling errors in column names
cleaned_tables["products"] = cleaned_tables["products"].rename(columns={
    "product_name_lenght": "product_name_length",
    "product_description_lenght": "product_description_length"
})
```

```python
# Convert designated date strings into actual datetime objects
date_columns = {
    "orders": [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ],
    "order_items": ["shipping_limit_date"],
    "order_reviews": ["review_creation_date", "review_answer_timestamp"],
    "closed_deals": ["won_date"],
    "marketing_qualified_leads": ["first_contact_date"]
}

for table_name, cols in date_columns.items():
    for col in cols:
        cleaned_tables[table_name][col] = pd.to_datetime(
            cleaned_tables[table_name][col],
            errors="coerce"
        )
```

```python
# Pad zip codes with leading zeros to maintain 5 digits
def clean_zip_prefix(series):
    """
    Converts zip code prefixes to 5-character strings.
    Useful for PostgreSQL and preserving zip-code style formatting.
    """
    return (
        series
        .astype("Int64")
        .astype("string")
        .str.zfill(5)
    )

zip_columns = {
    "customers": ["customer_zip_code_prefix"],
    "sellers": ["seller_zip_code_prefix"],
    "geolocation": ["geolocation_zip_code_prefix"]
}

for table_name, cols in zip_columns.items():
    for col in cols:
        cleaned_tables[table_name][col] = clean_zip_prefix(cleaned_tables[table_name][col])
```

```python
# Standardize state abbreviations (uppercase, remove spaces)
def clean_state_column(series):
    """
    Standardizes Brazilian state abbreviations.
    """
    return (
        series
        .astype("string")
        .str.strip()
        .str.upper()
    )

state_columns = {
    "customers": ["customer_state"],
    "sellers": ["seller_state"],
    "geolocation": ["geolocation_state"]
}

for table_name, cols in state_columns.items():
    for col in cols:
        cleaned_tables[table_name][col] = clean_state_column(cleaned_tables[table_name][col])
```

```python
# Standardize text columns (lowercase, strip, single spaces)
def clean_text_column(series):
    """
    Standardizes text values:
    - strips whitespace
    - converts to lowercase
    - replaces multiple spaces with single spaces
    """
    return (
        series
        .astype("string")
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", " ", regex=True)
    )

text_columns = {
    "customers": ["customer_city"],
    "sellers": ["seller_city"],
    "geolocation": ["geolocation_city"],

    "orders": ["order_status"],

    "order_payments": ["payment_type"],

    "products": ["product_category_name"],

    "product_category_translation": [
        "product_category_name",
        "product_category_name_english"
    ],

    "marketing_qualified_leads": [
        "landing_page_id",
        "origin"
    ],

    "closed_deals": [
        "business_segment",
        "lead_type",
        "lead_behaviour_profile",
        "average_stock",
        "business_type"
    ]
}

for table_name, cols in text_columns.items():
    for col in cols:
        cleaned_tables[table_name][col] = clean_text_column(cleaned_tables[table_name][col])
```

```python
# Fill missing categorical values with 'unknown'
cleaned_tables["marketing_qualified_leads"]["origin"] = (
    cleaned_tables["marketing_qualified_leads"]["origin"]
    .fillna("unknown")
)

closed_deal_unknown_cols = [
    "business_segment",
    "lead_type",
    "lead_behaviour_profile",
    "average_stock",
    "business_type"
]

for col in closed_deal_unknown_cols:
    cleaned_tables["closed_deals"][col] = cleaned_tables["closed_deals"][col].fillna("unknown")

cleaned_tables["products"]["product_category_name"] = (
    cleaned_tables["products"]["product_category_name"]
    .fillna("unknown")
)
```

```python
# Map string representations to true boolean data types
bool_cols = ["has_company", "has_gtin"]

for col in bool_cols:
    cleaned_tables["closed_deals"][col] = (
        cleaned_tables["closed_deals"][col]
        .astype("string")
        .str.strip()
        .str.lower()
        .map({
            "true": True,
            "false": False,
            "1": True,
            "0": False
        })
        .astype("boolean")
    )
```

```python
# Remove duplicate rows from the geolocation table
before_rows = len(cleaned_tables["geolocation"])

cleaned_tables["geolocation"] = (
    cleaned_tables["geolocation"]
    .drop_duplicates()
    .reset_index(drop=True)
)

after_rows = len(cleaned_tables["geolocation"])

print(f"Geolocation rows before removing duplicates: {before_rows}")
print(f"Geolocation rows after removing duplicates: {after_rows}")
print(f"Duplicate rows removed: {before_rows - after_rows}")
```

```python
# Sync product categories by appending missing ones to the translation table
products_categories = set(
    cleaned_tables["products"]["product_category_name"].dropna().unique()
)

translation_categories = set(
    cleaned_tables["product_category_translation"]["product_category_name"].dropna().unique()
)

# Find missing categories
missing_categories = sorted(products_categories - translation_categories)

# Create new dataframe rows
missing_category_rows = pd.DataFrame({
    "product_category_name": missing_categories,
    "product_category_name_english": missing_categories
})

# Append and drop potential duplicates
cleaned_tables["product_category_translation"] = pd.concat(
    [
        cleaned_tables["product_category_translation"],
        missing_category_rows
    ],
    ignore_index=True
).drop_duplicates(subset=["product_category_name"])

print("Missing product categories added to translation table:")
print(missing_categories)
```

```python
# Flag sellers in closed_deals based on their presence in sellers and orders tables
cleaned_tables["closed_deals"]["seller_exists_in_sellers"] = (
    cleaned_tables["closed_deals"]["seller_id"]
    .isin(cleaned_tables["sellers"]["seller_id"])
)

# Check if seller actually sold anything
converted_sellers_with_orders = set(cleaned_tables["order_items"]["seller_id"].unique())

cleaned_tables["closed_deals"]["seller_has_orders"] = (
    cleaned_tables["closed_deals"]["seller_id"]
    .isin(converted_sellers_with_orders)
)
```

```python
# Output a summary of table shapes and missing values post-cleaning
cleaning_summary = []

for name, df in cleaned_tables.items():
    cleaning_summary.append({
        "table": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "duplicate_rows": df.duplicated().sum(),
        "missing_values": df.isna().sum().sum()
    })

cleaning_summary_df = pd.DataFrame(cleaning_summary)
display(cleaning_summary_df)
```

```python
# Re-run foreign key check to ensure no new orphans were created
relationship_checks = [
    ("orders", "customer_id", "customers", "customer_id"),
    ("order_items", "order_id", "orders", "order_id"),
    ("order_items", "product_id", "products", "product_id"),
    ("order_items", "seller_id", "sellers", "seller_id"),
    ("order_payments", "order_id", "orders", "order_id"),
    ("order_reviews", "order_id", "orders", "order_id"),
    ("closed_deals", "mql_id", "marketing_qualified_leads", "mql_id"),
    ("closed_deals", "seller_id", "sellers", "seller_id"),
    ("products", "product_category_name", "product_category_translation", "product_category_name")
]

for left_table, left_key, right_table, right_key in relationship_checks:
    left_values = set(cleaned_tables[left_table][left_key].dropna())
    right_values = set(cleaned_tables[right_table][right_key].dropna())

    unmatched = left_values - right_values

    print(f"{left_table}.{left_key} → {right_table}.{right_key}")
    print("Unmatched values:", len(unmatched))
    print("-" * 60)
```

```python
# Extract year, month, and month_start from MQL contacts for analysis
mql = cleaned_tables["marketing_qualified_leads"]

mql["first_contact_year"] = mql["first_contact_date"].dt.year
mql["first_contact_month"] = mql["first_contact_date"].dt.month
mql["first_contact_month_start"] = mql["first_contact_date"].dt.to_period("M").dt.to_timestamp()

cleaned_tables["marketing_qualified_leads"] = mql
```

```python
# Calculate sales cycle length (days to close) and flag invalid cycles
closed_deals = cleaned_tables["closed_deals"]
mql_dates = cleaned_tables["marketing_qualified_leads"][["mql_id", "first_contact_date"]]

# Bring in contact dates
closed_deals = closed_deals.merge(
    mql_dates,
    on="mql_id",
    how="left"
)

closed_deals["won_year"] = closed_deals["won_date"].dt.year
closed_deals["won_month"] = closed_deals["won_date"].dt.month
closed_deals["won_month_start"] = closed_deals["won_date"].dt.to_period("M").dt.to_timestamp()

# Calculate cycle time in days
closed_deals["days_to_close"] = (
    closed_deals["won_date"] - closed_deals["first_contact_date"]
).dt.days

# Flag negative or invalid close times
closed_deals["is_valid_sales_cycle"] = closed_deals["days_to_close"] >= 0

cleaned_tables["closed_deals"] = closed_deals
```

```python
# Calculate delivery metrics (delays, estimated times, and late flags)
orders = cleaned_tables["orders"]

orders["purchase_year"] = orders["order_purchase_timestamp"].dt.year
orders["purchase_month"] = orders["order_purchase_timestamp"].dt.month
orders["purchase_month_start"] = orders["order_purchase_timestamp"].dt.to_period("M").dt.to_timestamp()

orders["approval_days"] = (
    orders["order_approved_at"] - orders["order_purchase_timestamp"]
).dt.total_seconds() / 86400

orders["delivery_days"] = (
    orders["order_delivered_customer_date"] - orders["order_purchase_timestamp"]
).dt.total_seconds() / 86400

orders["estimated_delivery_days"] = (
    orders["order_estimated_delivery_date"] - orders["order_purchase_timestamp"]
).dt.total_seconds() / 86400

# Delay calculation (actual - estimated)
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"] - orders["order_estimated_delivery_date"]
).dt.total_seconds() / 86400

# Flag if delivered past the estimated date
orders["is_late_delivery"] = orders["delivery_delay_days"] > 0

cleaned_tables["orders"] = orders
```

```python
# Add combined order value (price + freight)
order_items = cleaned_tables["order_items"]

order_items["item_total_value"] = order_items["price"] + order_items["freight_value"]

cleaned_tables["order_items"] = order_items
```

```python
# Calculate total product volume (L x W x H)
products = cleaned_tables["products"]

products["product_volume_cm3"] = (
    products["product_length_cm"]
    * products["product_height_cm"]
    * products["product_width_cm"]
)

cleaned_tables["products"] = products
```

```python
# Create flag for text reviews and calculate review response times
reviews = cleaned_tables["order_reviews"]

reviews["has_review_comment"] = reviews["review_comment_message"].notna()

reviews["review_response_time_hours"] = (
    reviews["review_answer_timestamp"] - reviews["review_creation_date"]
).dt.total_seconds() / 3600

cleaned_tables["order_reviews"] = reviews
```

```python
# Flag orders paid using multiple installments
payments = cleaned_tables["order_payments"]

payments["is_installment_payment"] = payments["payment_installments"] > 1

cleaned_tables["order_payments"] = payments
```

```python
# Output heads of the newly derived columns to visually verify
derived_columns_check = {
    "marketing_qualified_leads": [
        "first_contact_year",
        "first_contact_month",
        "first_contact_month_start"
    ],
    "closed_deals": [
        "first_contact_date",
        "won_year",
        "won_month",
        "won_month_start",
        "days_to_close",
        "is_valid_sales_cycle"
    ],
    "orders": [
        "purchase_year",
        "purchase_month",
        "purchase_month_start",
        "approval_days",
        "delivery_days",
        "estimated_delivery_days",
        "delivery_delay_days",
        "is_late_delivery"
    ],
    "order_items": [
        "item_total_value"
    ],
    "products": [
        "product_volume_cm3"
    ],
    "order_reviews": [
        "has_review_comment",
        "review_response_time_hours"
    ],
    "order_payments": [
        "is_installment_payment"
    ]
}

for table_name, cols in derived_columns_check.items():
    print("=" * 80)
    print(table_name)
    print("=" * 80)
    display(cleaned_tables[table_name][cols].head())
```

```python
# Locate anomalous calculated values (negatives or zeros)
print("Closed deals with negative days_to_close:")
display(
    cleaned_tables["closed_deals"]
    .loc[cleaned_tables["closed_deals"]["days_to_close"] < 0,
         ["mql_id", "first_contact_date", "won_date", "days_to_close"]]
)

print("Orders with negative delivery_days:")
display(
    cleaned_tables["orders"]
    .loc[cleaned_tables["orders"]["delivery_days"] < 0,
         ["order_id", "order_purchase_timestamp", "order_delivered_customer_date", "delivery_days"]]
)

print("Orders with negative approval_days:")
display(
    cleaned_tables["orders"]
    .loc[cleaned_tables["orders"]["approval_days"] < 0,
         ["order_id", "order_purchase_timestamp", "order_approved_at", "approval_days"]]
)

print("Products with zero or missing volume:")
display(
    cleaned_tables["products"]
    .loc[
        (cleaned_tables["products"]["product_volume_cm3"].isna()) |
        (cleaned_tables["products"]["product_volume_cm3"] <= 0),
        ["product_id", "product_length_cm", "product_height_cm", "product_width_cm", "product_volume_cm3"]
    ]
)
```

```python
# Output updated table summary showing new column counts
updated_summary = []

for name, df in cleaned_tables.items():
    updated_summary.append({
        "table": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "duplicate_rows": df.duplicated().sum(),
        "missing_values": df.isna().sum().sum()
    })

updated_summary_df = pd.DataFrame(updated_summary)
display(updated_summary_df)
```

```python
# Mask out negative sales cycles with pd.NA
closed_deals = cleaned_tables["closed_deals"]

closed_deals["is_valid_sales_cycle"] = closed_deals["days_to_close"] >= 0

# Replace negatives with NA
closed_deals.loc[
    closed_deals["days_to_close"] < 0,
    "days_to_close"
] = pd.NA

cleaned_tables["closed_deals"] = closed_deals
```

```python
# Convert `is_late_delivery` to nullable boolean and handle NAs
orders = cleaned_tables["orders"]

orders["is_late_delivery"] = pd.NA

orders.loc[
    orders["order_delivered_customer_date"].notna(),
    "is_late_delivery"
] = orders.loc[
    orders["order_delivered_customer_date"].notna(),
    "delivery_delay_days"
] > 0

orders["is_late_delivery"] = orders["is_late_delivery"].astype("boolean")

cleaned_tables["orders"] = orders
```

```python
# Verify previously detected anomalies were resolved
print("Invalid sales cycles:")
display(
    cleaned_tables["closed_deals"]
    .loc[
        cleaned_tables["closed_deals"]["is_valid_sales_cycle"] == False,
        ["mql_id", "first_contact_date", "won_date", "days_to_close", "is_valid_sales_cycle"]
    ]
)

print("Late delivery value counts:")
display(cleaned_tables["orders"]["is_late_delivery"].value_counts(dropna=False))

print("Product volume missing:")
display(
    cleaned_tables["products"]
    .loc[
        cleaned_tables["products"]["product_volume_cm3"].isna(),
        ["product_id", "product_volume_cm3"]
    ]
)
```
