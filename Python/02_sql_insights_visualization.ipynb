```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from matplotlib.ticker import FuncFormatter
```

```python
# Set display options for pandas DataFrames
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Define paths for the project
project_path = Path(
    r"D:\Data_Science\Projects\Portfolio_Projects\Marketing_Funnel_Seller_Performance_and_Revenue_Impact"
)

analysis_results_path = project_path / "Dataset" / "Analysis_Results"
visuals_path = project_path / "Assets" / "Visuals" 

visuals_path.mkdir(parents=True, exist_ok=True)
```

```python
# Load analysis results into a dictionary of DataFrames
result_files = {
    "q1_marketing_channel_business_value": "q1_marketing_channel_business_value.csv",
    "q2_landing_page_lead_quality": "q2_landing_page_lead_quality.csv",
    "q3_lead_profile_prioritization": "q3_lead_profile_prioritization.csv",
    "q4_sales_process_efficiency": "q4_sales_process_efficiency.csv",
    "q5_seller_activation_overall": "q5_seller_activation_overall.csv",
    "q5_seller_activation_by_channel": "q5_seller_activation_by_channel.csv",
    "q6_seller_group_marketplace_outcomes": "q6_seller_group_marketplace_outcomes.csv"
}

results = {}

for name, file_name in result_files.items():
    file_path = analysis_results_path / file_name
    results[name] = pd.read_csv(file_path)
    print(f"{name} loaded | Shape: {results[name].shape}")
```

```python
# Display the first few rows of each DataFrame
for name, df in results.items():
    print("=" * 80)
    print(name)
    print("=" * 80)
    display(df.head())
```

```python
# Create copies of the DataFrames for visualization
q1 = results["q1_marketing_channel_business_value"].copy()
q2 = results["q2_landing_page_lead_quality"].copy()
q3 = results["q3_lead_profile_prioritization"].copy()
q4 = results["q4_sales_process_efficiency"].copy()
q5_overall = results["q5_seller_activation_overall"].copy()
q5_channel = results["q5_seller_activation_by_channel"].copy()
q6 = results["q6_seller_group_marketplace_outcomes"].copy()
```

```python
# Data transformation for visualization
def clean_label(value):
    return str(value).replace("_", " ").title()

# Q1 / Q5 channel labels
q1["marketing_channel_label"] = q1["marketing_channel"].apply(clean_label)
q5_channel["marketing_channel_label"] = q5_channel["marketing_channel"].apply(clean_label)

# Q2 landing page short label
q2["landing_page_short"] = q2["landing_page_id"].str[:8] + "..."

# Q3 lead profile label
q3["lead_profile_label"] = (
    q3["lead_type"].apply(clean_label)
    + " | "
    + q3["business_segment"].apply(clean_label)
    + " | "
    + q3["lead_behaviour_profile"].apply(clean_label)
    + " | "
    + q3["business_type"].apply(clean_label)
)

# Q4 sales segment label
q4["sales_segment_label"] = (
    q4["marketing_channel"].apply(clean_label)
    + " | "
    + q4["lead_type"].apply(clean_label)
    + " | "
    + q4["business_segment"].apply(clean_label)
    + " | "
    + q4["business_type"].apply(clean_label)
)

# Q6 seller group label
q6["seller_group_label"] = (
    q6["business_segment"].apply(clean_label)
    + " | "
    + q6["lead_type"].apply(clean_label)
    + " | "
    + q6["business_type"].apply(clean_label)
)
```

```python
# Dark visual style settings
BACKGROUND_COLOR = "#111827"
AXES_COLOR = "#111827"
MAIN_COLOR = "#38BDF8"
TEXT_COLOR = "#F9FAFB"

plt.rcParams.update({
    "figure.facecolor": BACKGROUND_COLOR,
    "axes.facecolor": AXES_COLOR,
    "savefig.facecolor": BACKGROUND_COLOR,
    "axes.edgecolor": BACKGROUND_COLOR,
    "axes.labelcolor": TEXT_COLOR,
    "xtick.color": TEXT_COLOR,
    "ytick.color": TEXT_COLOR,
    "text.color": TEXT_COLOR,
    "axes.titleweight": "bold",
    "axes.titlesize": 16,
    "axes.labelsize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})
```

```python
# Formatting helper functions
def format_brl(x, pos):
    return f"R$ {x:,.0f}"

def format_percent(x, pos):
    return f"{x:.0f}%"

def clean_chart_style():
    ax = plt.gca()

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

    ax.tick_params(axis="x", bottom=False, labelbottom=False)
    ax.tick_params(axis="y", length=0)

    ax.grid(False)

    ax.set_xlabel("")
    ax.set_ylabel("")

def save_chart(file_name):
    output_path = visuals_path / file_name
    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Saved chart: {output_path}")
```

```python
# Q1: Total Seller Revenue by Marketing Channel
q1_revenue = q1.sort_values("total_seller_revenue", ascending=True)

plt.figure(figsize=(12, 6))

bars = plt.barh(
    q1_revenue["marketing_channel_label"],
    q1_revenue["total_seller_revenue"],
    color=MAIN_COLOR
)

plt.title("Total Seller Revenue by Marketing Channel", fontsize=16, pad=1)
plt.xlabel("Total Seller Revenue (BRL)", fontsize=12)
plt.ylabel("Marketing Channel", fontsize=12)

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_brl))

for bar in bars:
    width = bar.get_width()
    plt.text(
        width,
        bar.get_y() + bar.get_height() / 2,
        f" R$ {width:,.0f}",
        va="center",
        fontsize=10
    )

clean_chart_style()

save_chart("q1_total_seller_revenue_by_channel.png")
```

```python
# Q1: MQL-to-Deal Conversion Rate by Marketing Channel
q1_conversion = q1.sort_values("mql_to_deal_conversion_rate", ascending=True)

plt.figure(figsize=(12, 6))

bars = plt.barh(
    q1_conversion["marketing_channel_label"],
    q1_conversion["mql_to_deal_conversion_rate"],
    color=MAIN_COLOR
)

plt.title("MQL-to-Deal Conversion Rate by Marketing Channel", fontsize=16, pad=1)
plt.xlabel("Conversion Rate (%)", fontsize=12)
plt.ylabel("Marketing Channel", fontsize=12)

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_percent))

for bar in bars:
    width = bar.get_width()
    plt.text(
        width,
        bar.get_y() + bar.get_height() / 2,
        f" {width:.2f}%",
        va="center",
        fontsize=10
    )

clean_chart_style()

save_chart("q1_mql_to_deal_conversion_rate_by_channel.png")
```
